```python
import pyodbc
import numpy as np

# Connect to datasource
conn = pyodbc.connect('Trusted_Connection=yes', driver = '{SQL Server}',server = 'smc-sql-2014,1434\phi ', database = 'DataBridge_Ben')

# Create cursor associated with connection
cursor=conn.cursor()

# Query to sql database that cursor can execute on.
querystring = "SELECT TOP 1000 * FROM BBAB_ActivePersonDetail"

# Execute cursor on query string
cursor.execute(querystring)

## Get column names
columns = [column[0] for column in cursor.description]
```

```python
## Print column names
print(columns)
```

```python
## Use pandas to insert directly into a pandas dataframe
import pandas as pd

df = pd.read_sql(querystring,conn)
```

```python
df[0:4]
```

```python
#Show rows with only Female, non-graduates who got loans.  
#Similar to a WHERE clause.
df.loc[(df["GenderDescription"]=="Female") 
         & (df["Blind"]==0) 
         & (df["PrimaryLanguageDesc"]=="English"), ["GenderDescription","Blind","PrimaryLanguageDesc"]]
```

```python
## Use numpy to create and Excel-style pivot table
```

```python
#Determine pivot table
impute_grps = df.pivot_table(values=["HouseholdAnnualIncome"], index=["GenderDescription","AgeRange","EducationDesc"], aggfunc=np.mean)
print(impute_grps)
```
